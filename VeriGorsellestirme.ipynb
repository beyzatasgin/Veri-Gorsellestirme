```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Veri setini yükle (dosya yolunu kendi sisteminize göre güncelleyin)
df = pd.read_csv("train_users_2.csv")
```

```python
df['date_account_created'] = pd.to_datetime(df['date_account_created'], errors='coerce')
df['year'] = df['date_account_created'].dt.year
```

```python
##  1. Çubuk Grafik (matplotlib) - En Çok Rezervasyon Yapılan Ülkeler
```

```python
df['country_destination'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("En Çok Rezervasyon Yapılan 10 Ülke")
plt.xlabel("Ülke")
plt.ylabel("Kullanıcı Sayısı")
plt.show()
```

```python
##2. Çizgi Grafik (pyplot) - Yıllara Göre Kullanıcı Kayıtları
```

```python
df['year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Yıllara Göre Kullanıcı Kayıtları")
plt.xlabel("Yıl")
plt.ylabel("Kayıt Sayısı")
plt.show()
```

```python
## # 3. Pasta Grafik (matplotlib) - Cinsiyet Dağılımı
```

```python
df['gender'].replace('-unknown-', 'Unknown').value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Kullanıcıların Cinsiyet Dağılımı")
plt.ylabel("")
plt.show()
```

```python
# 4. Histogram (seaborn) - Kullanıcı Yaş Dağılımı
```

```python
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Kullanıcı Yaş Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Frekans")
plt.show()
```

```python
# 5. Dağılım Grafiği (plotly) - Yaş ve Kayıt Yılı İlişkisi
```

```python
fig = px.scatter(df, x="age", y="year", color="gender",
                 title="Yaş ve Yıl Bazında Kullanıcı Dağılımı",
                 labels={"age":"Yaş", "year":"Kayıt Yılı"})
fig.show()
```
